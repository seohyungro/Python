```python
kospi_3 = kospi_3[['회사명', '거래소코드', '회계년도', '정상영업이익증가율', '순이익증가율', '매출액증가율']]
```

```python
kospi_3['회계년도'] = kospi_3['회계년도'].astype(str)
kospi3['회계년도'] = kospi3['회계년도'].astype(str)
kospi_3['회사명'] = kospi_3['회사명'].astype(str)
kospi3['회사명'] = kospi3['회사명'].astype(str)
kospi_3['거래소코드'] = kospi_3['거래소코드'].astype(str)
kospi3['거래소코드'] = kospi3['거래소코드'].astype(str)
```

```python
kospi_3.info()
```

```python
kospi3.info()
```

```python
k_3 = pd.DataFrame(columns=kospi_3.columns)  # 빈 데이터프레임 생성
k_3_ = pd.DataFrame(columns=kospi3.columns)  # 빈 데이터프레임 생성

for i in range(len(kospi_3)):
    for j in range(len(kospi3)):
        if (kospi_3.loc[i, '회사명'] == kospi3.loc[j, '회사명'] and
            kospi_3.loc[i, '거래소코드'] == kospi3.loc[j, '거래소코드'] and
            kospi_3.loc[i, '회계년도'] == kospi3.loc[j, '회계년도']):
            k_3 = pd.concat([k_3, kospi_3.loc[[i]]], ignore_index=True)
            k_3_ = pd.concat([k_3_, kospi3.loc[[j]]], ignore_index=True)
            break
```

```python
k_3
```

```python
kospi3 = pd.concat([k_3[['회사명', '거래소코드', '회계년도', '정상영업이익증가율', '순이익증가율', '매출액증가율']], k_3_[['주당순손익', '부채비율', '매출액정상영업이익률']]], axis=1)
```
