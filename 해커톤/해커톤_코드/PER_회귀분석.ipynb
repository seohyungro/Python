```python
indi = 'PER'
```

```python
import pandas as pd
import numpy as np

kospi = pd.read_csv("./kospi(18~22).csv")
kospi = kospi[['회사명', '회계년도', indi, '수익률']]
# kospi['NEW'] = -1*kospi['NEW']
kospi
```

```python
import statsmodels.api as sm

X = kospi[indi]
y = kospi['수익률']

X = sm.add_constant(X)

model = sm.OLS(y, X)

results = model.fit()

print(results.summary())
```
