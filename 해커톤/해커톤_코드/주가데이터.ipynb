```python
import pandas as pd
import numpy as np
```

```python
kospi = pd.read_csv('kospi.csv')
corp_names = kospi['회사명'].unique().tolist()
corp_names = [name.replace('(주)', '') for name in corp_names]
corps = pd.DataFrame(corp_names, columns=['corp_name'])
corps
```

```python
corp_codes = pd.read_csv('corp_codes.csv')
```

```python

for corp_name in corps['corp_name']:
    filtered_rows = corp_codes[corp_codes['corp_name']==corp_name]
    if not filtered_rows.empty:
        corps.loc[corps['corp_name']==corp_name, 'corp_code'] = filtered_rows['stock_code'].values[0]
```
