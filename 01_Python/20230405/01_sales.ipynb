### 연습문제
1. csv 폴더에 있는 Sales Records.csv 파일 로드 
2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel' 컬럼을 삭제
3. 컬럼의 이름을 변경 -> ['권역', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
4. 권역과 국가를 기준으로 오름차순 정렬
5. index를 초기화하고 기존의 인덱스는 삭제

```python
# 파일 로드 
# 상위 폴더로 이동 -> csv 폴더로 이동 -> 파일명
sales = pd.read_csv("../csv/Sales Records.csv")
sales.head(2) # 체크
```

```python
# Order ID 컬럼을 삭제
sales = sales.drop("Order ID", axis=1)
```

```python
sales
```

```python
# 남은 4개 컬럼 동시에 삭제
#하나의 매개변수에 집어넣기 위해서 리스트로 묶음 (리스트나 튜플이나 딕셔너리 하나로 묶기 가능)
sales.drop(["Sales Channel", "Total Revenue", "Total Cost", "Total Profit"], axis=1, inplace=True)
```

```python
sales
```

```python
# 컬럼의 이름을 변경
sales.columns = ['권역', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
```

```python
sales.head(5)
```

```python
# 권역, 국가를 기준으로 오름차순 정렬
sales.sort_values(["권역", "국가"], inplace=True)
```

```python
sales
```

```python
# 인덱스를 초기화하고 기존의 인덱스를 삭제
sales.reset_index(drop=True, inplace=True)
```

```python
sales.head()
```

- 새로운 파생 변수 생성
    1. 총단가 --> 판매단위 * 단가
    2. 총원가 --> 판매단위 * 원가
    3. 총이윤 --> 총단가 - 총원가

```python
sales['총단가'] = sales['판매단위'] * sales['단가']

# 시리즈 데이터프레임 공부!
```

```python
sales['총원가'] = sales['판매단위'] * sales['원가']
```

```python
# 총이윤 -> 총단가 - 총원가
sales['총이윤'] = sales['총단가'] - sales['총원가']
```
