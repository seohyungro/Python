- 새로운 파생 변수 생성
    1. 총단가 --> 판매단위 * 단가
    2. 총원가 --> 판매단위 * 원가
    3. 총이윤 --> 총단가 - 총원가

```python
df['총단가'] = df['판매단위']*df['단가']
```

```python
df
```

```python
df['총원가'] = df['판매단위'] * df['원가']
df['총이윤'] = df['총단가'] - df['총원가']
```
