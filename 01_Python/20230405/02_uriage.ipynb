1. 상품A의 item_price의 값은 결측치 중 상품A이면 이 값을 치환
2. 나머지 상품들도 같은 작없을 반복

```python
# 상품A의 item_price를 출력
# 행의 조건? 결측치가 아니다, item_name이 상품 A이다,
# 열의 조건? item_price

# 결측치가 아닌 조건
flg_null = uriage['item_price'].isna()
# uriage.isna().any(axis=1) # 위에랑 결과는 같은데 속도가 더 느림
# ~ 사용하면 반대
```

```python
# 결측치인 값들을 출력
uriage.loc[flg_null] # 행에 적기(열에 적을려면 uriage.loc[ , flg_null])
```

```python
# item_name이 상품A인 경우만 출력
uriage.loc[uriage['item_name'] == '상품A']

# 결측치는 변수로 넣고 이건 변수로 안만들어 놓은 이유는 상품A를 계속 바꿔야되고 결측치는 안바뀌어두되서
```

```python
# 상품A이고 item_price가 결측치인 경우
uriage.loc[(flg_null) & (uriage['item_name'] == '상품A')]
```

```python
# 상품A이고 item_price가 결측치가 아닌 경우
price = uriage.loc[(uriage['item_name'] == '상품A') & ~(flg_null), 'item_price'].value_counts().index[0]
```

```python
# 결측치에 값을 치환하는 함수
# fillna(n) : n의 값으로 결측치를 채운다

uriage.loc[(flg_null) & (uriage['item_name'] == '상품A')].fillna(price)
```

```python
uriage.loc[(flg_null) & 
    (uriage['item_name'] == '상품A'), 'item_price'] = price
```

```python
# A~Z까지 해야되서 for문에 넣어서 하기
# 위에는 A만 한거

# item_name의 리스트를 출력
name_list = uriage['item_name'].unique()
```

```python
for i in name_list:
    # item_price의 값을 구한다
    # item_name이 i와 같고 결측치가 아닌 데이터에서 price를 출력
    # ~(flg_null) = 결측치가 아닌값
    price = uriage.loc[~(flg_null) & \
        (uriage['item_name'] == i), 'item_price']\
            .value_counts().index[0]
    # item_name이 i와 같고 결측치인 데이터에서 item_price에 price라는 변술을 삽입
    # & 로 연결할 때 앞뒤에 () 쓰기 ex) () & ()

    uriage.loc[(flg_null) & \
        (uriage['item_name'] == i), 'item_price'] = price
    # \ 는 한줄에 적어야하는데 밑줄에 내려가서 이어지지않을때 이어지게 해주는 표현
    # apply랑 lambda 같이?
```
