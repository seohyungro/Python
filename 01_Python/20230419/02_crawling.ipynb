```python
# 크롤링할때 필요한 라이브러리
import requests
from bs4 import BeautifulSoup as bs
```

```python
# 크롤링 할때 get()사용 : 주어진 URL에서 HTML 페이지를 가져옵니다
requests.get('https://finance.naver.com/marketindex/exchangeList.naver')
# 결과 <Response [200]> : 정상적이다
```

```python
# 실제 사이트에서 html 정보를 받아와서 변수에 저장
data = requests.get('https://finance.naver.com/marketindex/exchangeList.naver').text
```

```python
# BeautifulSoup을 이용하여 데이터를 파싱
soup = bs(data, 'html.parser')
```

```python
print(soup)
```

```python
# find는 html안에서 해당 태그의 첫번째 정보를 출력
tbody = soup.find('tbody')
```

```python
# find_all() 해당 태그의 전체 정보를 리스트의 형태로 출력
soup.find_all('td')[0].get_text().replace('\n','').strip('\t')
# soup.find_all('td')[0].get_text().replace('\n','').replace('\t','')
```

```python
# find_all() 해당 태그의 전체 정보를 리스트의 형태로 출력
data_list = soup.find_all('td')
```

```python
# data_list를 이용하여 2차원 배열로 새로운 리스트를 생성
# 1차원 리스트 생성
# [[7개씩]], [[7개씩]], [[7개씩]], [[7개씩]] -> 2차원 리스트
```

```python
# 1차원 리스트 생성
_list =[]
for i in data_list:
    x = i.get_text()
    x = x.replace('\n','')
    x = x.strip('\t')
    _list.append(x)
_list
```

```python
len(_list)//7
```

```python
for i in range(0, (len(_list)//7), 1):
    x = []
    for j in range(i*7, (i+1)*7, 1):
        x.append(j)
    print(x)
```

```python
# 2차원으로 나옴
result = []
for i in range(0, (len(_list)//7), 1):
    x = []
    for j in range(i*7, (i+1)*7, 1):
        x.append(_list[j])
    result.append(x)
result
```

```python
# DB에 table 생성 (naver)
# 필드는 7개 모두 필드의 타입은 문자열로 구성
# 통화명 필드는 primary key로 지정

sql ="""
    create
    table
    `naver`(
        `통화명` varchar(16),
        `매매기준율` varchar(16),
        `현찰(구매)` varchar(16),
        `현찰(판매)` varchar(16),
        `송금(보낼때)` varchar(16),
        `송금(받을때)` varchar(16),
        `미화환산율` varchar(16)
    )
"""
db.sql_query(sql)
```

```python
# naver라는 이름으로 생성한 테이블에 데이터 삽입
# 데이터를 삽입하는 insert문을 작상
sql = """
    insert
    into
    naver
    values(
        '통화명',
        '1',
        '2',
        '3',
        '4',
        '5',
        '6'
    )
"""

db.sql_query(sql)
```
