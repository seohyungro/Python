```python
import pandas as pd
import numpy as np

adjust_price = pd.read_csv("./kospi(18~22).csv")
adjust_price
```

```python
adjust_price['전년 수익률'] = adjust_price['수익률'].shift()
adjust_price.dropna(inplace=True)
adjust_price.head(15)
```

```python
import statsmodels.api as sm

X = adjust_price['전년 수익률']
y = adjust_price['수익률']

X = sm.add_constant(X)

model = sm.OLS(y, X)

results = model.fit()

print(results.summary())
```
