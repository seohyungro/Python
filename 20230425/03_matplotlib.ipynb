```python
import platform

if platform.system() == 'Darwin':
    plt.rc('font', family = 'AppleGothic')
else:
    plt.rc('font', family = 'Malgun Gothic')
```

```python
# 그래프 사이즈를 조절
plt.figure(figsize=(14,6))
# grid 추가
plt.grid(True, axis = 'y')
plt.plot(kospi['Close'].head(20), marker ='o')
plt.xlabel('날짜')
plt.ylabel('종가')
plt.show()
```

```python
mean_kospi_2020 = kospi.loc['2020'].resample('MS').mean()
mean_kospi_2020
```

```python
# x축은 'XXXX년XX월' 표시
mean_kospi_2020.index[0].strftime('%Y년%m월')
```

```python
x = []
for i in mean_kospi_2020.index:
    x.append(i.strftime('%Y년%m월'))

x
```

```python
list(map(lambda x : x.strftime('%Y년%m월'), mean_kospi_2020.index))
```

```python
# 가로그래프 : barh
plt.barh(x, mean_kospi_2020['Volume'])
# 지수표현식을 일반 표현식으로 변경
plt.ticklabel_format(axis='x', style='plain')
plt.show()
```

```python
# 세로 그래프 : bar
plt.bar(x, mean_kospi_2020['Volume'])
# 지수표현식을 일반 표현식으로 변경
plt.ticklabel_format(axis='y', style='plain')
plt.show()
```

```python
# kospi에서 변동률 파생변수
# 변동률 = (종가-시가) / 시가 * 100
kospi['변동률'] = (kospi['Close'] - kospi['Open']) / kospi['Open'] * 100
kospi.head()
```

```python
# 음수를 표시
plt.rc('axes', unicode_minus =  False)
plt.grid()
plt.hist(kospi['변동률'], bins=20)
plt.show()
```

# 문제
1. test csv 폴더 안에 있는 오프라인.csv를 로드
2. 컬럼의 이름을 ['기준년월', '업종', '성별', '연령대', '총이용건수', '총이용금액']으로 변경
3. 기준년월을 시계열 데이터로 변경
4. index를 기준년월로 변경
5. 온라인.csv 파일도 위의 작업을 반복한다.
6. 온라인에 업종 컬럼을 추가하여 '온라인'으로 값을 지정
7. 두개의 데이터프레임을 유니얼결합
8. 업종별로 총이용금액의 합계를 그래프로 표시

```python
import pandas as pd
```

```python
offline = pd.read_csv('../test_csv/유통+업종(백화점,+할인점,+슈퍼마켓,+편의점)+카드소비트렌드+데이터_1901_2109.csv', encoding="euc-kr")
```

```python
offline
```

```python
offline.columns = ['기준년월', '업종', '성별', '연령대', '총이용건수', '총이용금액']
```

```python
offline['기준년월'] = pd.to_datetime(offline['기준년월'], format="%Y%m")
```

```python
offline = offline.set_index('기준년월')
```

```python
offline
```

```python
online = pd.read_csv('../test_csv/온라인+카드소비트렌드+데이터_1901_2109.csv', encoding= 'euc-kr')
```

```python
online
```

```python
online['업종'] = '온라인'
```

```python
online
```

```python
online.columns = ['기준년월', '성별', '연령대', '총이용건수', '총이용금액', '업종']
```

```python
online
```

```python
online['기준년월'] = pd.to_datetime(online['기준년월'], format="%Y%m")
online.info()
```

```python
online.set_index('기준년월', inplace=True)
```

```python
data = pd.concat([offline, online], axis = 0)
```

```python
# 인덱스를 기준으로 오름차순 정렬
data.sort_index(inplace=True)
```

```python
pivot_data = pd.pivot_table(
    data,
    index = data.index,
    columns = '업종',
    values = '총이용금액',
    aggfunc = 'sum'
)
```

```python
plt.rc('font', family = 'AppleGothic')
```

```python
pivot_data.plot()
```

```python
group_data = data.groupby([data.index, '업종']).sum()['총이용금액']
```

```python
group_data.unstack().plot()
```

1. 결합된 데이터프레임에서 성별을 1은 '남자' 2는 '여자'로 변경
2. 온라인 업종의 남, 여의 카드매출 월별로 어떻게 차이가 나는가?를 시각화하여 표시

```python
data
```

```python
data['성별'].map({1:'남자', 2:'여자'})
```

```python
data['성별'] = data['성별'].apply(lambda x: '남자' if x == 1 else '여자')  # 조건식에서 == 시용
```

```python
data.loc[data['성별'] == 1,['성별']] = '남자'
data.loc[data['성별'] == 2,['성별']] = '여자'
data['성별'].value_counts
```

```python
online_data = data.loc[data['업종'] == '온라인']
```

```python
pd.pivot_table(
    online_data,
    index = online_data.index,
    columns = '성별',
    values=  '총이용금액',
    aggfunc = 'mean'
).plot()
```
